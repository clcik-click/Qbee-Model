# Experiment 1 - General Knowledge

### Configure

```python
import os
print(os.getcwd())
```

```python
# Loads the model
from ultralytics import YOLO
model = YOLO("yolo11n.pt")
```

### Training

```python
# Trains 
model.train(
    data="exp1.yaml",
    imgsz=896,
    epochs=100,
    batch=16,
    patience=20,
    device="mps", 
    workers=0      
)
```

### Testing on test_30

```python
model = YOLO("../runs/detect/train/weights/best.pt")
metrics = model.val(
    data="exp1.yaml",
    split="test",
    device="mps",
    workers=0
)
```
